```python
import seaborn as sns
import metapack as mp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display 

%matplotlib inline
sns.set_context('notebook')
mp.jupyter.init()
```

```python
#pkg = mp.jupyter.open_package()
pkg = mp.jupyter.open_source_package()
pkg
```

```python
links = pkg.reference('ca_sd_links').dataframe().set_index('geoid')

links.head()
```

```python
from collections import defaultdict
frames = defaultdict(list)

for r in pkg.references():
    if r.get_value('tableid'):
        df = r.dataframe().titles

        # Bug is public data package -- spaces after names
        df.columns = [c.strip() for c in df.columns]
        df = df.drop(columns = ['STUSAB', 'COUNTY', 'NAME'])
        frames[r.get_value('tableid')].append(df)

    
```

```python
frames = { k:pd.concat(v) for k,v in frames.items() }
```

```python
df = links
for f in frames.values():
    df = df.join(f)
```
